# Analysis of Euro Exchange Rates (1999–2021)

The dataset used in this analysis describes **Euro daily exchange rates** from **1999 to 2021**. The **Euro (€)** is the official currency of most **European Union** countries.  

This dataset was compiled by **Daria Chemkaeva** and made available on **Kaggle**. The data source is the **European Central Bank**, and it receives regular updates. We downloaded the dataset in **January 2021**.  

**Dataset link:** [Euro Exchange Daily Rates (1999–2020) on Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

The dataset contains **5,699 rows** and **41 columns**. Most columns have missing values, and their data type is predominantly **object**. Only three columns are of type **float64**.

## Data Cleaning  

To ensure the dataset is well-structured and ready for analysis, we performed the following cleaning steps:  

- **Renaming Columns:** We renamed certain columns for clarity.   
- **Converting Data Types:** The `'Time'` column was originally of type **object**. We converted it to **datetime** format for proper time-based analysis.  
- **Sorting Data:** We sorted the dataset by the `'Time'` column in chronological order.  
- **Resetting Index:** After sorting, we reset the index to maintain a clean and structured format.  

These preprocessing steps improve data accessibility and ensure consistency for further analysis.  

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]       
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

### Handling Missing and Invalid Values in `'US_dollar'`  

The `'US_dollar'` column contained **62 rows** with a `'-'` value, which represents missing data. To ensure accuracy in our analysis, we performed the following steps:  

- **Removing Invalid Values:** We dropped all rows where `'US_dollar'` had a `'-'` value.  
- **Converting Data Type:** After cleaning, we converted the `'US_dollar'` column from **object** to **float** for numerical analysis.  

These steps ensure that the exchange rate data is clean and ready for further exploration.  


```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
```

```python
euro_to_dollar.loc[:, 'US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
%matplotlib inline
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
```

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

### Smoothing Exchange Rates with Rolling Mean  

To focus on long-term trends in the exchange rates, we applied a **rolling mean** with a **window size of 30 days**. This technique helps smooth out short-term fluctuations and highlights broader patterns over time.  

By averaging the values over a 30-day period, we gain a clearer view of long-term changes in the **US dollar exchange rate**.  


```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(12, 6))

plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

```python
euro_to_dollar_2006_2010 = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2006, 2009)]
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(16, 8))

plt.plot(euro_to_dollar_2006_2010['Time'],
         euro_to_dollar_2006_2010['rolling_mean'],
         c = 'lightgreen')

tick_positions = [1.2, 1.3, 1.4, 1.5]
ax.set_yticks(tick_positions)

selected_years = [2006, 2007, 2008, 2009, 2010]
selected_ticks = pd.to_datetime([f"{year}-01-01" for year in selected_years])

ax.set_xticks(selected_ticks)  
ax.set_xticklabels(selected_years) 

mask_2007_2009 = euro_to_dollar_2006_2010["Time"].dt.year.between(2007, 2008)
ax.plot(euro_to_dollar_2006_2010["Time"][mask_2007_2009], 
        euro_to_dollar_2006_2010["rolling_mean"][mask_2007_2009], 
        color="red")

ax.axvspan(xmin=13955, xmax=14125, ymin=0.09,
           alpha=0.3, color='grey')


ax.text(13149.0, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold', size=24)
ax.text(13149.0, 1.65, 'Euro-USD exchange rates between 2006 and 2010',
        size=16)

ax.grid(zorder=0)

x_min, x_max = ax.get_xlim() 
y_min = ax.get_ylim()[0] - 0.05

ax.text(x_min, y_min, 
        "© Nurbek Khujaev".ljust(165) + "Source: European Central Bank",
        color='#f0f0f0', backgroundcolor='#4d4d4d', 
        fontsize=14, ha="left", zorder=3)

ax.grid(alpha=0.5)

plt.show()
```
